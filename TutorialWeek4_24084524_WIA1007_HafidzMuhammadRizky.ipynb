##### Question #1: Does Work/Study for more than or equal to 6 hours and sleeping less correlate with depression?

```python
data_comparison_1 = data[(data['Work/Study Hours'] >= 6) & ((data['Sleep Duration'] == "'Less than 5 hours'") | (data['Sleep Duration'] == "'5-6 hours'") )]
data_comparison_1
```

```python
data_comparison_2 = data[(data['Work/Study Hours'] >= 6) & ((data['Sleep Duration'] == "'7-8 hours") | (data['Sleep Duration'] == "'More than 8 hours'") )]
data_comparison_2
```

```python
total_1 = len(data_comparison_1)
total_2 = len(data_comparison_2)
depressed_1_count = data_comparison_1['Depression'].eq(1).sum()
depressed_2_count = data_comparison_2['Depression'].eq(1).sum()
percent_1 = (depressed_1_count / total_1) * 100
percent_2 = (depressed_2_count / total_2) * 100
print(f"Depression Rate (1): {percent_1:.2f}%")
print(f"Depression Rate (2): {percent_2:.2f}%")
```

##### Question #2: Are unhealthy dietary habits linked to higher depression rates?

```python
data_comparison_1_dietary = data[(data['Dietary Habits'] == 'Healthy') | (data['Dietary Habits'] == 'Moderate')]
data_comparison_1_dietary
```

```python
data_comparison_2_dietary = data[(data['Dietary Habits'] == 'Unhealthy')]
data_comparison_2_dietary
```

```python
total_1 = len(data_comparison_1_dietary)
total_2 = len(data_comparison_2_dietary)
depressed_1_dietary_count = data_comparison_1_dietary['Depression'].eq(1).sum()
depressed_2_dietary_count = data_comparison_2_dietary['Depression'].eq(1).sum()
percent_1 = (depressed_1_dietary_count / total_1) * 100
percent_2 = (depressed_2_dietary_count / total_2) * 100
print(f"Depression Rate Dietary (1): {percent_1:.2f}%")
print(f"Depression Rate Dietary (2): {percent_2:.2f}%")
```
