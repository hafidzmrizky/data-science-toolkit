```python
import numpy as np
```

```python
# 1. Numpy Array of 2x2
npArr1 = np.array([[1,2], [3,4]])
print(npArr1)
print("Shape of:", npArr1.shape)
```

```python
# 2. Random Numpy Array with size of 3x4
npArr2 = np.random.random((3,4))
print(npArr2)
print("Shape of:",npArr2.shape)
```

```python
# 3. Numpy Array of ones and zeros
npArr3 = np.ones((2,2))
npArr4 = np.zeros((2,2))
print(npArr3)
print("Shape of ones:", npArr3.shape)
print(npArr4)
print("Shape of zeros:", npArr4.shape)
```

```python
# 4. Random numpy arr of size 6x3, then transpose
npArr5 = np.random.random((6,3))
print("Normal:\n", npArr5)
npArr5_T = npArr5.T
print("Transposed:\n", npArr5_T)
```

```python
# 5. Two random numpy array size of n x p, and p x m, then compute the dot product
n = np.random.randint(2,5)
p = np.random.randint(2,5)
m = np.random.randint(2,5)

npArr6 = np.random.random((n,p))
npArr7 = np.random.random((p,m))
print("First array:\n", npArr6, "\n Shape of first Array: ", npArr6.shape)
print("Second array:\n", npArr7, "\n Shape of second Array: ", npArr7.shape)

npArr8 = np.dot(npArr6, npArr7)
print("Dot product:\n", npArr8)
```
